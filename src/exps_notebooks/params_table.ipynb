```python
import pandas as pd
import yaml
```

```python
with open(r'../configs/ue/mc.yaml') as file:
    mc = yaml.load(file, Loader=yaml.FullLoader)

with open(r'../configs/ue/mc-dpp.yaml') as file:
    mc_dpp = yaml.load(file, Loader=yaml.FullLoader)
    
with open(r'../configs/ue/mc_mahalanobis.yaml') as file:
    mc_mahalanobis = yaml.load(file, Loader=yaml.FullLoader)
```

```python
dpp_params = ['committee_size', {'dropout': ['max_frac', 'committee_size', 'mask_name', 'mask_name_for_mask']}]
mc_params = ['committee_size', 'inference_prob']
mc_mahalanobis_params = ['committee_size']
```

```python
mc_mahalanobis_params_df = pd.DataFrame()
for key in mc_mahalanobis_params:
    mc_mahalanobis_params_df[key] = [mc_mahalanobis[key]]

dpp_params_df = pd.DataFrame()
for key in dpp_params:
    if isinstance(key, dict):
        for key_ in key:
            for val in key[key_]:
                dpp_params_df[key_+'_'+val] = mc_dpp[key_][val]
    else:
        dpp_params_df[key] = [mc_dpp[key]]

mc_params_df = pd.DataFrame()
for key in mc_params:
    mc_params_df[key] = [mc[key]]
```
