```python
import pandas as pd
import os


def format_arc_table_results(results, baseline_col, subtract=False, percents=False, ndp=2):
    if subtract:
        baseline = results.T[baseline_col]
        diff_res = results.T.drop(columns=baseline_col).subtract(baseline, axis="rows")
        diff_res = pd.concat([baseline], axis=1).T
    else:
        diff_res = results

    mean_res = results.groupby(level=0).mean()
    std_res = results.groupby(level=0).std()
    if percents:
        mean_res *= 100
        std_res *= 100
    formatted_results = mean_res.applymap(
        lambda x: "{}±".format(round(x, ndp))
    ) + std_res.applymap(lambda x: "{}".format(round(x, ndp)))

    return formatted_results


def load_table(path):
    with open(path, 'r') as f:
        file = f.read()
    table = pd.read_json(file)
    table = table.set_index('index')
    return table
```

```python
# Set pathes to saved metrics tables
base_path = '/home/jovyan/uncertainty-estimation/workdir/run_calc_ues_metrics/mixup_xlnet_test/amazon/msd/last'
path_macro = os.path.join(base_path, 'metrics_table_f1_macro.json')
path_micro = os.path.join(base_path, 'metrics_table_f1_micro.json')

f1_macro = load_table(path_macro)
f1_micro = load_table(path_micro)
```

```python
f1_micro_tab = format_arc_table_results(f1_micro, 'max_prob', ndp=3)
f1_micro_tab
```
